```python
from PIL import Image, ImageDraw, ImageFont
import textwrap

# Set the DPI and dimensions
DPI = 300
WIDTH = int(2.5 * DPI)
HEIGHT = int(3.5 * DPI)
BORDER = int(0.15 * DPI)
# Calculate dimensions in pixels

main_image = Image.open("games/risk_2025/img/input/action1_bushtroops.jpg")

# Resize the image to fit card width - border, keeping the aspect ratio
main_image = main_image.resize(
    (
        WIDTH - 2 * BORDER,
        int(main_image.size[1] * (WIDTH - 2 * BORDER) / main_image.size[0]),
    )
)

# # Crop image to fit half the card height - border
# dracula = dracula.crop((0, 0, dracula.size[0], HEIGHT // 2 - BORDER))

# Create new blank image
card = Image.new("RGBA", (WIDTH, HEIGHT), (0, 0, 0, 255))
draw = ImageDraw.Draw(card)

# Set y axis positions for key elements
TITLE_Y = BORDER + 20
IMAGE_Y = TITLE_Y + 50
MAIN_TEXT_Y = IMAGE_Y + main_image.size[1] + 20
QUOTE_TEXT_Y = MAIN_TEXT_Y + 20


card.paste(main_image, (BORDER, IMAGE_Y))


title = "Project For A New American Century"
title_font = ImageFont.truetype("fonts/Anton-Regular.ttf", 48)

main_text = "Choose a territory another player controls. You may move any number of units from that territory to another that player controls, leaving a minimum of one in place."
main_text = textwrap.fill(main_text, width=45)
main_text_font = ImageFont.truetype("fonts/Anton-Regular.ttf", 36)

quote_text = "If you want a serious interrogation, you send a prisoner to Jordan. If you want them to be tortured, you send them to Syria. If you want someone to disappear  never to see them again  you send them to Egypt."
quote_text_font = ImageFont.truetype("fonts/Anton-Regular.ttf", 24)

# Draw the title text
draw.text(
    (WIDTH // 2, BORDER + 20),
    title,
    font=title_font,
    fill="white",
    align="center",
    anchor="mm",
)

# Draw the main text
draw.text(
    (BORDER, MAIN_TEXT_Y),
    main_text,
    font=main_text_font,
    fill="white",
    align="left",
)
# # Draw the quote text
# draw.text(
#     (BORDER, QUOTE_TEXT_Y),
#     quote_text,
#     font=quote_text_font,
#     fill="white",
#     align="left",
# )
card
```
